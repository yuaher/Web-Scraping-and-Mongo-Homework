```python
#Import dependencies
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
from splinter import Browser
```

```python
# Initializing Chromedriver for browser used
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#URL of page to be scraped: NASA Mars News site
url = 'https://mars.nasa.gov/news/'

# Visit url
browser.visit(url)

# Retrieve page with the requests module
response = browser.html

# Creating BeautifulSoup object and parse with 'html.parser'
soup = BeautifulSoup(response, 'html.parser')

# Extract text for news title & description
title = soup.find('div', 'content_title', 'a').text
news_p = soup.find('div', 'article_teaser_body').text


# Print latest news title & description
print(title)
print(news_p)
```

```python
# Retrieve parent divs for content titles
article_title = soup.find_all('div', 'content_title')

# Create empty list for content titles
article_list = []

# Loop through content titles
for article in article_title:
    # Add each title to article_list
    article_list.append(article.find('a').text)
    # Display all article titles
    print(article.find('a').text)
    print('----------------')
```

```python
# Setting up splinter browser
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser('chrome', **executable_path, headless=False)

# Setting variable for Mars Facts Webpage
url = 'https://space-facts.com/mars/'

# Visit url
browser.visit(url)

# Retrieve page with the requests module
response = browser.html

# BeautifulSoup object and parse with 'html.parser'
soup = BeautifulSoup(response, 'html.parser')

# Find and show entire table
table= soup.find('table', 'tablepress tablepress-id-mars')
print(table.prettify())
```

```python
# Retrieving all the rows in the table
rows = table.find_all('tr')

# Creating empty lists to hold td elements from each row
labels = []
values = []

# Looping through rows & add first 'td' element to labels list and values list
for row in rows:
    td_elements = row.find_all('td')
    print(td_elements[0].text)
    print(td_elements[1].text)
    print('------------------------------')
    labels.append(td_elements[0].text)
    values.append(td_elements[1].text)
```

```python
# Create Dataframe
mars_data = pd.DataFrame({" ": labels,
                           "Values": values})

# Set a new index
mars_data.set_index(" ", inplace=True)

# Show Dataframe
mars_data
```

```python
# Splinter browser
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser('chrome', **executable_path, headless=False)

# Setting variable for Mars Facts Webpage
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'

# Visiting the url
browser.visit(url)

# Retrieve page with the requests module
response = browser.html

# Creating BeautifulSoup object and parse with 'html.parser'
soup = BeautifulSoup(response, 'html.parser')

# Retrieving class for each hemisphere available
returns = soup.find('div', 'collapsible results')

# Retrieving all anchors in each class pulled above
hemispheres = returns.find_all('div', {"class": 'description'})

# Creating empty list to hold dictionaries
hemi_img_urls = []

# Looping through all anchors for each hemisphere class
for description in hemispheres:
    a = description.find('a')
    
    # Retrieving title and link to specific hemisphere page
    hemi_title = a.h3.text
    hemi_link = 'https://astrogeology.usgs.gov' + a['href']
    
    # Visiting the above link
    browser.visit(hemi_link)
    time.sleep(1)

    #Retrieving the link to image
    page = browser.html
    results = BeautifulSoup(page, 'html.parser')
    image_link = results.find('div', 'downloads').find('li').a['href']

    # Creating empty dictionary for title & image
    image_dict = {}
    image_dict['title'] = hemi_title
    image_dict['img_url'] = image_link

    # Adding image_dict to empty list from above
    hemi_img_urls.append(image_dict)
    
print(hemi_img_urls)

# Converting results to a dataframe for display
test = pd.DataFrame(hemi_img_urls, columns= ["title","img_url"])
test
```
